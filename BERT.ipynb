# Modeling

In BERT uncased, the text has been lowercased before WordPiece tokenization step while in BERT cased, the text is same as the input text (no changes).

```python
from simpletransformers.classification import ClassificationModel

train_args = {
    "evaluate_during_training": True,
    "evaluate_during_training_verbose": True,
    "tensorboard_dir": "logs",
}

model = ClassificationModel('bert', 'bert-base-cased', args={'fp16': False, 'num_train_epochs': 3, "overwrite_output_dir": True})
```

```python
model.train_model(train, eval_df=validation)
```

```python
# Evaluate the model and get probabilities
import sklearn
result, model_outputs, wrong_predictions = model.eval_model(test, acc=sklearn.metrics.accuracy_score, f1=sklearn.metrics.f1_score, output_dir="outputs/")
```

```python
from scipy.special import softmax

# Get both predictions and raw outputs
predictions, raw_outputs = model.predict(test_data["text"].tolist())

# Convert the raw outputs (logits) to probabilities using softmax
probabilities = softmax(raw_outputs, axis=1)

# Now you can iterate over the probabilities and print them out
for i, (pred, prob) in enumerate(zip(predictions, probabilities)):
    print(f"Text {i}:")
    print(f"Predicted class: {'Human' if pred == 0 else 'Machine'}")
    print(f"Probability of being human: {prob[0]*100:.2f}%")
    print(f"Probability of being machine: {prob[1]*100:.2f}%")
```

# Insights on how model working

```python
import pandas as pd
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
type_error_ratio = new_predict.groupby('type').apply(lambda group: 1.0 - accuracy_score(group['label'], group['predictions']))
print("Error Ratio by Type:")
print(type_error_ratio)

# Calculate error ratio for every source
source_error_ratio = new_predict.groupby('source').apply(lambda group: 1.0 - accuracy_score(group['label'], group['predictions']))
print("\nError Ratio by Source:")
print(source_error_ratio)

# Calculate error ratio for every topic
topic_error_ratio = new_predict.groupby('topic').apply(lambda group: 1.0 - accuracy_score(group['label'], group['predictions']))
print("\nError Ratio by Topic:")
print(topic_error_ratio)
```

```python
ax = type_error_ratio.plot(kind='bar', x='type', y='value', legend=False)
ax.set_ylabel('error ratio')
ax.set_title('Type Error Ratio')
# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
ax = source_error_ratio.plot(kind='bar', x='source', y='value', legend=False)
ax.set_ylabel('error ratio')
ax.set_title('source error ratio')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

plt.show()
```
