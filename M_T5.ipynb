```python
import sys
train_data_path = "/content/drive/MyDrive/Graduation_Project_Ottawa/datasets/arabic and english dataset/en_ar_train.csv"
test_data_path="/content/drive/MyDrive/Graduation_Project_Ottawa/datasets/arabic and english dataset/en_ar_test.csv"
# sys.path.insert(0, train_data)

random_state = 523
```

```python
import csv
import pandas as pd
train_data= pd.read_csv(train_data_path)
test_data= pd.read_csv(test_data_path)
```

```python
train=pd.DataFrame(columns=['prefix','input_text','target_text'])
test=pd.DataFrame(columns=['prefix','input_text','target_text'])
```

```python
train
```

```python

train['input_text']=train_data['text']
train['target_text']=train_data['label']
test['input_text']=test_data['text']
test['target_text']=test_data['label']
```

```python
train['prefix'] = "classification"
test['prefix'] = "classification"
```

```python
# train['input_text'] = train['input_text'].astype(str)
# test['input_text'] = test['input_text'].astype(str)
```

```python
train
```

```python
# from simpletransformers.classification import ClassificationModel
from simpletransformers.t5 import T5Model
# Create a ClassificationModel with mT5 model
model = T5Model(
    'mt5',
    'google/mt5-base',  # You can use other mT5 variants as needed
    args={'fp16': False, 'num_train_epochs': 3, 'overwrite_output_dir': True}
)
```

```python
train_data = pd.DataFrame({
    'prefix': train['prefix'],
    'input_text': train['input_text'].astype(str),  # Ensure input_text is in string format
    'target_text': train['target_text'].astype(str)  # Ensure target_text is in string format
})
```

```python
train_data
```

```python
model.train_model(train_data)
```

```python
test_data = pd.DataFrame({
    'prefix': test['prefix'],
    'input_text': test['input_text'].astype(str),  # Ensure input_text is in string format
    'target_text': test['target_text'].astype(str)  # Ensure target_text is in string format
})
```

```python
test_data
```

```python

import sklearn
result = model.eval_model(test_data)
```

```python
result
```

```python
predict=model.predict(test_data['input_text'].tolist())
```

```python
predictions1=pd.DataFrame(predict, columns=['column_name'])
```

```python
from sklearn.metrics import accuracy_score, f1_score
accuracy = accuracy_score(test_data['target_text'].astype(int), predictions1.astype(int))
f1 = f1_score(test_data['target_text'].astype(int), predictions1.astype(int))
```
